```python
import pandas as pd
import os

# 1. Tentukan path ke file Anda. Ini sudah benar.
file_path = os.path.join('..', 'dataset', 'dataset_label.json')

# 2. Baca file JSON menggunakan pandas untuk membuat DataFrame.
# Ini adalah langkah yang hilang dari kode Anda.
df = pd.read_json(file_path)

# 3. Sekarang df adalah DataFrame, dan baris kode selanjutnya akan berjalan dengan benar.
df['label_numeric'] = df['label'].apply(lambda x: 1 if x == 'hate_speech' else 0)

# 4. Tampilkan jumlah masing-masing label.
print(df['label'].value_counts())
```

```python
# Tentukan nama untuk file yang sudah diproses
processed_file_path = 'dataset_processed.csv'

# Simpan DataFrame ke file CSV
# index=False penting agar indeks DataFrame (0, 1, 2, ...) tidak ikut disimpan sebagai kolom baru
df.to_csv(processed_file_path, index=False)

print(f"DataFrame berhasil disimpan di: {processed_file_path}")
```
